## Question 3

How many rows are there for the Yellow Taxi data for all CSV files in the year 2020?

-----

I didn't exactly ingest all the tables one by one into Big Query. I used the ```GCSToBigQueryOperator()``` from Airflow to send everything to the same partitioned table. When running this query:

```
SELECT COUNT(*) FROM `dez-bootcamp.yellow_taxi.yellow_trips`
WHERE DATE(tpep_pickup_datetime) BETWEEN DATE("2020-01-01") AND DATE("2020-12-31")
```

The result of this query is 24,649,725, which doesn't match with any of the answers, but it's pretty close to 24,648,499. What I'm assumming is happening, is that there are some rows from the 2022 and 2020 .csvs which were fed into the 2021 .csv. Since the numbers are so close, this is my final answer.

**Final Answer: 24,648,499**
